```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression,SGDClassifier
```

```python
height = float(input("Enter your height: "))
weight = float(input("Enter your weight: "))
age = int(input("Enter your age: "))
gender = input("Enter your gender: ").upper()
clas = input("Enter your body type: ").upper()
situps = input("Enter the number of situps you can do: ")
conditions = input("Enter if you have any health issues(heart/hormonal/lungs/nil): ")
first_filter = file[(file["height_cm"]>height-3) & (file["height_cm"]<height+3) & (file["weight_kg"]>weight-3) & (file["weight_kg"]<weight+5)]
second_filter = first_filter[(first_filter['age']>age-5) & (first_filter["age"]<age+5)]
third_filter = second_filter[(second_filter["gender"] == gender) & (second_filter["class"] == clas) & (file["health"] == conditions)]
print(third_filter)
if len(third_filter) == 0:
    print("Couldn't find workouts")
elif len(third_filter) == 1:
    print(str(third_filter["workouts"]))
elif len(third_filter) <15:
    lab = LabelEncoder()
    stan = StandardScaler()
    sca = StandardScaler()
    tran = lab.fit_transform(third_filter["workouts"])
    X = third_filter[["bicep_improvement","leg_improvement","chest_improvement","shoulder_improvement","sit-ups counts","weight_kg","height_cm"]]
    stoic = SGDClassifier()
    stoic.fit(stan.fit_transform(X),tran)
    x = np.array([third_filter["bicep_improvement"].mean(),third_filter["leg_improvement"].mean(),third_filter["chest_improvement"].mean(),third_filter["shoulder_improvement"].mean(),situps,weight,height]).reshape(1,-1)
    print(lab.inverse_transform(stoic.predict(stan.fit_transform(x))))
else:
    lab = LabelEncoder()
    sca = StandardScaler()
    tran = lab.fit_transform(third_filter["workouts"])
    X = third_filter[["bicep_improvement","leg_improvement","chest_improvement","shoulder_improvement","sit-ups counts","weight_kg","height_cm"]]
    X_train,X_test,y_train,y_test = train_test_split(X,tran)
    s_train = sca.fit_transform(X_train)
    s_test = sca.fit_transform(X_test)
    dec = DecisionTreeClassifier()
    ran = RandomForestClassifier()
    ada = AdaBoostClassifier()
    lg = LogisticRegression()
    sgd = SGDClassifier()
    sv = SVC()
    dec.fit(s_train,y_train)
    ran.fit(s_train,y_train)
    ada.fit(s_train,y_train)
    lg.fit(s_train,y_train)
    sgd.fit(s_train,y_train)
    sv.fit(s_train,y_train)
    dec_pred = dec.predict(s_test)
    sv_pred = sv.predict(s_test)
    ran_pred = ran.predict(s_test)
    ada_pred = ada.predict(s_test)
    lg_pred = lg.predict(s_test)
    sg_pred = sgd.predict(s_test)
    final = {"Decision":accuracy_score(dec_pred,y_test),"Random":accuracy_score(ran_pred,y_test),"AdaBoost":accuracy_score(ada_pred,y_test),"Logistic":accuracy_score(lg_pred,y_test),"Stoichastic":accuracy_score(sg_pred,y_test)}
    x = np.array([third_filter["bicep_improvement"].mode(),third_filter["leg_improvement"].mode(),third_filter["chest_improvement"].mode(),third_filter["shoulder_improvement"].mode(),situps,weight,height]).reshape(1,-1)
    print(final)
    if max(final.keys()) == "Stoichastic":
        sgd.fit(sca.fit_transform(X),tran)
        print(lab.inverse_transform(sgd.predict(sca.fit_transform(x))))
    elif max(final.keys()) == "Decision":
        dec.fit(sca.fit_transform(X),tran)
        print(lab.inverse_transform(dec.predict(sca.fit_transform(x))))
    elif max(final.keys()) == "Random":
        ran.fit(sca.fit_transform(X),tran)
        print(lab.inverse_transform(ran.predict(sca.fit_transform(x))))
    elif max(final.keys()) == "AdaBoost":
        ada.fit(sca.fit_transform(X),tran)
        print(lab.inverse_transform(ada.predict(sca.fit_transform(x))))
    elif max(final.keys()) == "Logistic":
        lg.git(sca.fit_transform(X),tran)
        print(lab.inverse_transform(lg.predict(sca.fit_transform(x))))
```
